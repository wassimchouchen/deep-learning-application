```python
from pandas import read_excel
pd.set_option('display.max_columns', None)
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd 
import numpy as np
from re import sub
from pandas import DataFrame
class Processing():
    def __init__(self,data,tfidf,bow):
        self.bow = bow
        self.data_first = data # Data source
        self.tfidf = tfidf # model TF-IDF Vectorizer
        self.data_second = [] # Data Vectorized
        self.data_final = [] # Data similarity
   
    def itemDefinise(self):
        data = self.data_first# load data
        ID=data["Customer ID"].unique()#identfy ID
        ID.tolist()# convert to list
        ID.sort()# sorted ID
        item=data.groupby("Customer ID")["Product Description"].apply(list)
        items=[]
        for i in range(data["Customer ID"].nunique()):
            items.append(item.iloc[i])
        basket=DataFrame({"ticket_id":ID,"products":items})
        basket['products'] = basket['products'].astype(str)
        regex = r'[\[\]/,]'
        basket['products'] = basket['products'].apply(lambda x: sub(regex, '', x))
        basket['products'] = basket['products'].apply(lambda x: sub(r'(\w+)\s(\w+)', r'\1_\2', x))
        self.item =[ i.split(" ") for i in basket['products']]
        self.data_first = basket   
        return self.data_first #Data transformation
    def tfidf_vect(self):
        data = self.data_first
        self.data_second = self.tfidf.fit_transform(data['products']).toarray()
        return self.data_second
    def bow_vec(self):
        data = self.data_first
        bag_of_word = self.bow.fit_transform(data['products']).toarray()
        return bag_of_word
    def similarity(self):
        from sklearn.metrics.pairwise import cosine_similarity
        data = self.data_second
        self.data_final = cosine_similarity(data.T)
        return self.data_final
```

```python
data = read_excel('/kaggle/input/basket-analysis/Basket Analysis - Data.xlsx')
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
bow=CountVectorizer()
processing = Processing(data,tfidf,bow)
df = processing.itemDefinise()
tfidf_matrix=processing.tfidf_vect()
bow_matrix=processing.bow_vec()
similarity=processing.similarity()
```
