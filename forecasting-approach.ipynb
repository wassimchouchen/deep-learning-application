```python
import numpy as np
import pandas as pd 
```

```python
train=pd.read_csv("/kaggle/input/rossmann-store-sales/train.csv")
store=pd.read_csv("/kaggle/input/rossmann-store-sales/store.csv")
```

```python
train.head(2)
```

```python
train.tail(2)
```

```python
train.DayOfWeek.nunique(),train.Date.nunique(),train.Store.nunique()
```

```python
store["PromoInterval"]=store["PromoInterval"].fillna("Jan,Apr,Jul,Oct")
store.head()
```

```python
train=pd.merge(train,store, on="Store" )
train
```

```python
train=train.sample(frac=1)
train.reset_index(drop=True)
train=train.fillna(train.mode())
train
```

```python
train.to_csv("SalesForecasting.csv", index=False)
```
